Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. Создайте датафреймы X и y из этих данных.

Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки

составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.

Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.
Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.


```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
print(boston.keys())
data = boston["data"]
data.shape
```

```python
feature_names = boston['feature_names']
target = boston['target']
```

```python
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
X.info()
```

```python
y = pd.DataFrame(target, columns=["price"])
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

y_pred.shape
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head(10)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(check_test['y_test'], check_test['y_pred'])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
DATA_PATH = pd.read_csv('creditcard.csv')
df = pd.DataFrame(DATA_PATH)
```

```python
df.value_counts(normalize=True)
df.info()
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
X = df.drop(['Class'], axis=1)
y = pd.Series(df['Class'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
parameters = [{
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
clf.best_score_
```

```python
y_pred_proba = clf.predict_proba(X_test)
y_pred_proba
```

```python
y_pred_proba = y_pred_proba[:, 1]
y_pred_proba[:5]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
